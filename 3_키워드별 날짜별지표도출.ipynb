```python
# 감성지수 포함 데이터 파일
import pandas as pd

df = pd.read_csv('senti_score_final.csv', encoding='utf-8')
df
```

```python
# 대면모임, 온라인모임, 방역, 규제, 코로나 각 태그별 csv 파일 만들기
```

```python
is_moim = df['태그'] == '대면모임'
moim = df[is_moim]

moim
```

```python
grouped = moim['감성지수'].groupby(moim['날짜'])
b = dict(grouped.count())
day_count = b.values()
a = dict(grouped.mean())
nalja = a.keys()
mean_senti = a.values()
new = pd.DataFrame()
new['날짜'] = nalja
new['날짜카운팅'] = day_count
new['태그'] = '대면모임'
new['감성점수평균'] = mean_senti
new['Indicator'] = new['날짜카운팅'] * new['감성점수평균']
new
```

```python
new.to_csv('senti_score_moim.csv', encoding='utf-8-sig')
```
