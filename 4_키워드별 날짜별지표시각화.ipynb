```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='NanumGothic') # For Windows
df = pd.read_csv('senti_score_moim.csv', encoding='utf-8')
plt.plot(df['날짜'], df['Indicator'])
plt.xlabel('날짜')
plt.ylabel('지표')
plt.title('<대면모임의견지표>')

plt.show()
```
